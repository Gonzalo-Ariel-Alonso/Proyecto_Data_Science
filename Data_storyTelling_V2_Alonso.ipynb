# Por que y como empezar a invertir dinero
Bienvenidos! en este documento voy a contarles las ventajas y desventajas en conjuento con los riesgos y beneficios de invertir dinero, basandome en la experiencia y los datos de inverores conocidos y no tan conocidos. Pero antes debemos tener en claro algunos conceptos:

## ¿Que es invertir?
Si nos guiamos por la definicion de diccionario, la inversión es acto de no consumir recursos para satisfacer necesidades en el presente, sino de destinarlos a satisfacer necesidades en el futuro. Es decir, comprar algo que sabemos que no necesitaremos en la inmediates, sino que especulamos a que pueda darnos una mejor utilidad en el futuro.

## ¿Que es reserva de valor?
Antes de continuar debemos entender una definicion muy importante: la reserva de valor. Que un activo sea de reserva de valor significa que su valor precenta una variacion nula (idealmente) con respecto a otros activos del mercado o tienen una utilidad que se prolonga en el tiempo. 
A lo largo de la historia de la humanidad han existido varios activos que se utilizaron como reserva de valor, siendo el oro y la plata los mas conocidos por su escases y utilidad.


## ¿Que es el dinero?
Podriamos tener una larga conversacion de que es el dinero y su historia, pero no es el topico que nos compete, asi que nos limitaremos a definir cuales son sus funciones:

 - Ser un medio de intercambio: Permite a las personas comprar y vender bienes y servicios sin necesidad de utilizar el trueque directo, lo que facilita las transacciones comerciales.

- Servir como unidad de cuenta: Debe poder utilizarse como medida común para expresar el valor de bienes, servicios y activos. Permite comparar y valorar distintos bienes en términos de su valor monetario.

- **Ser una reserva de valor**

## ¿Que es la inflacion?
La inflacion suele ser definida MUY erroneamente como la suba generalizada de precios de los productos del mercado, algo que en realidad es una concecuencia y no una causa, ya que la definicion correcta de inflacion es la perdida de valor adquisitivo del dinero en el tiempo. Nuevamente nos encontramos con un tema en el que podriamos indagar mucho mas, pero tampoco es el topico de este documento.

Los que pertenecemos a paises altamente inflacionarios como venezuela o argentina sabemos en carne propia lo que es que los billetes de nuestro bolsillo valgan cada vez menos.
Observemos el siguiente grafico donde comparamos el valor del peso argentino contra el dolar estadounidense en los ultimos 10 años.


*grafico del dolar CCL contra el peso argentino en escala logaritmica - tradingview*

Podemos observar que a inicios del 2011 comprar un dolar costaba 4 pesos, mientras que a la fecha actual (Julio del 2023) comprar un dolar cuesta 546 pesos, eso significa una perdida del valor adquisitivo del..... 100 - 4 / 546 * 100% = **99.2% !!!!** Asi como lo ves, el peso argentino esta muy lejos de cumplir el rol de reserva de valor.

## Entonces... ¿el dolar es reserva de valor?
Uno podria pensar que la moneda mas utilizada globalmente sirve como reserva de valor, hasta yo mismo lo asumi en el grafico anterior al compararlo con la moneda que circula en mi pais, pero lamentablemente el dolar, al igual que el peso argentino, tambien sufre la inflacion. Una inflacion que parece casi imperseptible en el corto plazo, pero observemos la inflacion del dolar en los ultimos 50 años

```python
import pandas as pd
us_inflation = pd.read_csv('https://media.githubusercontent.com/media/Gonzalo-Ariel-Alonso/Proyecto_Data_Science/master/us_inflation.csv')
us_inflation['year'] = pd.to_datetime(us_inflation.iloc[:,0]).dt.year
us_inflation = us_inflation.iloc[:,1:3].groupby('year').mean()
us_inflation
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(14, 10))
ax.stackplot(us_inflation.index , us_inflation.iloc[:,0])
plt.xticks(list(range(us_inflation.index[0],us_inflation.index[-1],5))+[2021], rotation = 45)
plt.yticks(range(0,280,10))
ax.grid(alpha = 0.3)
plt.show()
```

Podemos observar que en los ultimos 100 años el dolar acumulo una inflacion del mas del 260%, si repetimos la cuenta que hicimos para calcular la perdida de valor de un dolar nos queda 100 - 1 / 2.6 * 100% = 61.5% de perdida de valor.
Si observamos el grafico, notaremos una aceleracion inflacionara a partir de la decada del 70, esto no es casualidad, ya que en 1971 el dolar abandono el patron oro, el cual garantizaba que cada dolar estaba respaldado por su costo equivalente en oro. 

# Por que debemos empezar invertir nuestro dinero

## Conservar el valor
Analizando los graficos anteriormente dados podemos llegar a la conclucion de que **en la actualidad no hay moneda de cambio que nos cubra de la inflacion.** Es por esto que el motivo mas basico de inverir es porque es la unica forma de **conservar el valor de nuestro capital.**

## Generar un interes
Como mencionamos en la definicion de invertir, es poner a disposicion recursos que no necesitamos en el precente, esto no es gratis, puesto que el tiempo tambien es un recurso muy valioso, es por esto que invirtiendo no solo podemos conservar el valor de nuestro dinero, sino que tambien que podemos obtener un porcentaje de interes que nos dara una ganancia proporcional a la cantidad de tiempo que dejemos nuestros activos congelados.

## Invertir como servicio al mercado
Al invertir nuestro dinero, ya sea en una accion o en un bono, estamos dando nuestro boto de confianza a la entidad emisora de la accion o el bono, eso significa que la misma utilizara este capital para expandirse, lo que permitira un mayor nivel de empleo y mayor capacidad de desarollo empresarial, las empresas mas grandes del mundo no podria llegar a todos los paises que llegan si no fuera por la cooperacion del mercado. Le aseguro que si usted observa de que marca estan echos todos los muebles, electrodomesticos y materiales que hay dentro de su casa, se encotrara con decenas de marcas de empresas que cotizan en bolsa. 

# Por que *NO* deberiamos empezar invertir nuestro dinero
Un detalle importante a destacar es frenar a aquellos que se hacercan al mundo de las inversiones por los motivos totalmente erroneos.

## Buscar dinero facil y rapido
Generar intereses con nuestro dinero no significa que puedes hacerte millonario de un dia para el otro, conseguir rentabilidades altas en el mercado implica un mayor riesgo a perdidas, aquellos que buscan sacar mucho dinero en poco tiempo con una baja inversion mas temprano que tarde terminan perdiendolo todo. Educarse sin parar es una parte clave para no perder todo nuestro capital y ser consumidos por la avaricia. 

##  Invertir sin ahorros ni ingresos
Mucha gente se hacerca al mercado con la intencios de vivir de las inversiones o de generar una fuente de ingresos, ambas cosas mencionadas son posibles, pero para esto primero se debe tener un capital de respaldo, definimos invertir como entregar recrusos del precente que no necesitamos en el corto plazo, invertir dinero que sabre que necesitare el proximo mes para pagar mis cuentas es una pesima practica, siempre ten en cuenta que del ahorro nace la inversion, no al revez. 

## Endeudarse para invertir
Si ya de por si invertir dinero que necestiaremos en el coro plaso es una mala practica, es mil veces peor utilizar dinero que no poseemos, tomar deuda no solo no nos genera ningun interes sino siempre tendremos que pagar mas de lo que nos prestaron. Si la intencion es invertir para un emprendimiento puede que si se requiera un monto inicial para empezar, pero en el caso de la bolsa o el mercado de bonos solo se necesitan unos pocos dolares para comprar una accion, aqui no hay excusa.

## Apalancarse sin experiencia alguna
El apalancamiento es una herramienta que se utiliza para multiplicar ganancias o multiplicar las perdidas, aunque muchos que dicen ser "expertos" siempre por algun motivo omiten esta ultima parte, solo existe una diferencia entre el apalancamiento y tomar deuda, y es que en caso de tener perdidas cuando estas apalancado no le deberas dinero a nadie, pero el capital que al no estar apalancado lo perderias en años, al si estarlo lo perderias en segundos, y no estoy exagerando.

# En que invertir
Tenga en cuenta que nadie que no sea asesor financiero puede darle una recomendacion de inversion, esto atenta directamente contra la ley. Lo que si puede hacer es intruirse sobre que activos se ajustan a su modelo de inversor.

## Modelos de inversor
Podriamos dividr la cantidad de modelos de inversor de forma infinita practicamente, pero desde un punto de vista practico se puede definir en tres tipos:

- Inverosres con mucha adversion al riesgo: Son aquellos que invierten en activos que generan un bajo rendimiento pero tiene una baja probabilidad de tener perdidas.
- Inverosres con adversion al riesgo moderada: Inverores dispuestos arriesgarse a la posibilidad de estar en perdida en el corto plazo, a cambio de generar rendimientos mas altos en el largo plazo.
- Inverosres con adversion al riesgo nula: Aquellos inversores que no temen a la posibilidad de perder todo o gran parte del capital invertido para generar rentabilidades muy altas en poco tiempo.

## ¿Que modelo inversor me sienta mejor?
Esta pregunta solo podra responderla usted luego de llevar varios años invirtiendo de forma constante, pero como orientacion basica existen dos factores importantes que se deben tener en cuenta:

- **Edad:** si usted es joven y se encuentra en la decada de sus 20 años puede darse el lujo de tomar inversiones moderadamente arriesgadas a largo plazo (mas de 20 años), pero si usted ya tiene mas de 70 años, salvo que quiera dejarle una gran herencia a sus hijos, seguramente quiera disfruta del fruto de sus inversiones, en tal caso le conviene mantener un perfil bajo y no arriesgarse a perder todo lo que consigui ahorrar durante su vida. 

- **Importancia del capital que invierte:** Es usted quien debe asignar si una determinada cantidad de dinero es mucha o es poca, no tendria sentido hablar de cifras, ya que 500 dolares para mi puede ser mucho pero para usted poco, todo dependera de que tanto valore determianda cantidad de dinero, por ende mientras mas crea usted que invertir X cantidad de dinero es un valor alto, mas conservadora debe ser la inversion, y viseversa, si usted cree que perder esa cantidad X no le significara un golpe mayor, puede darse el lujo de probar una inversion un poco mas riesgosa. 


Es importante tener en cuenta que **toda inversion conlleva un riesgo**, ya sea alto o bajo, el riesgo siempre esta ahi y debe ser contemplado en todo momento, una frase muy comunmente dicho que seguramente habran escuchado es "no poner todos los huevos en una sola canasta", esto significa que no es conveniente invertir solo en un activo, a mayor diversificacion habra menos posibilidades de que las perdidas puedan hacernos un daño significativo.
